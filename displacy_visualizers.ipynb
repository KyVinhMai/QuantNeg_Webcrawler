```python
import spacy
from spacy import displacy
import en_core_web_sm
from spacy.matcher import Matcher
nlp = en_core_web_sm.load()
```

```python
matcher = Matcher(nlp.vocab)
pattern = [{"LIKE_EMAIL": True}]
matcher.add("EMAIL_ADDRESS", [pattern])
doc = nlp("This is an email address: wmattingly@aol.com")
matches = matcher(doc)
matches
doc[6:7]
```

```python
doc = nlp("Every member of the minority in this case was not appointed by Presidents Reagan or Bush")
print(doc[1].n_rights)
print([child for child in doc[1].children])
displacy.render(doc, style="dep",jupyter=True)
```

```python
doc = nlp("every one of these organizations who have endorsed you did not agree with everything you did or every word you've spoken")
print(doc[7].is_sent_end)
displacy.render(doc, style="dep",jupyter=True)
```
